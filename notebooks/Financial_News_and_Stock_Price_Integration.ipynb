**Load and Prepare News Data**

```python
# Construct the file path
file_path = os.path.join(rpath, 'data', 'raw_news_data','raw_analyst_ratings.csv')

# Read the CSV file into a DataFrame
sentiment_data = pd.read_csv(file_path)
sentiment_data = sentiment_data[['date','headline','url','publisher','stock']]
sentiment_data.head()
```

```python
import re
timezone_pattern = re.compile(r'[\+\-]\d{2}:\d{2}')

# Apply the pattern to identify rows with timezone information
has_timezone = sentiment_data['date'].apply(lambda x: bool(timezone_pattern.search(str(x))) if pd.notnull(x) else False)

# Split into two DataFrames
sentiment_data_with_timezone = sentiment_data[has_timezone]
sentiment_data_without_timezone = sentiment_data[~has_timezone]

# Reset index for clarity (optional)
sentiment_data_with_timezone.reset_index(drop=True, inplace=True)
sentiment_data_without_timezone.reset_index(drop=True, inplace=True)

sentiment_data_with_timezone['date'] = pd.to_datetime(sentiment_data_with_timezone['date'], errors='coerce')
sentiment_data_with_timezone['date'] = sentiment_data_with_timezone['date'].dt.date

sentiment_data_without_timezone['date'] = pd.to_datetime(sentiment_data_without_timezone['date'], errors='coerce')
sentiment_data_without_timezone['date'] = sentiment_data_without_timezone['date'].dt.date

concatenated_sentiment_data = pd.concat([sentiment_data_with_timezone, sentiment_data_without_timezone], ignore_index=True)
concatenated_sentiment_data.sort_values(by='date', inplace=True)
concatenated_sentiment_data.reset_index(drop=True, inplace=True)

concatenated_sentiment_data
```

```python
concatenated_sentiment_data.isnull().sum()
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
# Calculate the sentiment of the headlines
concatenated_sentiment_data['sentiment'] = concatenated_sentiment_data['headline'].apply(lambda x: sia.polarity_scores(text=x)['compound'])
concatenated_sentiment_data['sentiment_category'] = pd.cut(concatenated_sentiment_data['sentiment'], bins=[-1, -0.5, -0.0001, 0.5, 1], labels=['Very Negative', 'Negative', 'Neutral', 'Positive'])
concatenated_sentiment_data['Date'] = pd.to_datetime(concatenated_sentiment_data['date'])
concatenated_sentiment_data.head()
```

**Correlation Analysis**

```python
from scipy.stats import pearsonr
import pandas as pd

# Assuming you have already loaded your DataFrames into these variables
dataframes = {
    'AAPL': data,
    'NVDA': NVDA_data,
    'AMZN': AMZN_data,
    'GOOG': GOOG_data,
    'TSLA': TSLA_data,
}

correlation_results = {}

# Iterate through each stock
for stock, df in dataframes.items():
    # Group by Date and calculate the mean of sentiment
    sentiment_df = concatenated_sentiment_data[concatenated_sentiment_data['stock'] == stock]
    sentiment_df = sentiment_df.groupby('Date', as_index=False)['sentiment'].mean()

    # Ensure the Date column in both DataFrames is of Datetime type
    df['Date'] = pd.to_datetime(df['Date'])

    # Calculate daily stock returns
    df['daily_return'] = df['Close'].pct_change()

    # Merge sentiment data with stock data
    merged_df = pd.merge(df, sentiment_df, on='Date', how='inner')

    # Drop rows with NaN values (which occur due to pct_change)
    merged_df.dropna(subset=['daily_return', 'sentiment'], inplace=True)

    # Calculate the Pearson correlation coefficient
    sentiment_return_corr, p_value = pearsonr(merged_df['sentiment'], merged_df['daily_return'])
    sentiment_closing_corr, p_value = pearsonr(merged_df['sentiment'], merged_df['Close'])
    
    # Store results
    correlation_results[stock] = {
        "sentiment_return_corr": sentiment_return_corr,
        "sentiment_closing_corr": sentiment_closing_corr
    }

# Print the correlation results
for stock, correlation in correlation_results.items():
    print(f"Stock: {stock}")
    print(f'Pearson correlation coefficient for sentiment and daily return: {correlation["sentiment_return_corr"]:.4f}')
    print(f'Pearson correlation coefficient for sentiment and closing price: {correlation["sentiment_closing_corr"]:.4f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Create a grid of subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 10))
axes = axes.flatten()  # Flatten the 2D array of axes to easily index

# Iterate through each stock
for i, (stock, df) in enumerate(dataframes.items()):
    # Group by Date and calculate the mean of sentiment
    sentiment_df = concatenated_sentiment_data[concatenated_sentiment_data['stock'] == stock]
    sentiment_df = sentiment_df.groupby('Date', as_index=False)['sentiment'].mean()

    # Merge sentiment data with stock data
    df['Date'] = pd.to_datetime(df['Date'])
    merged_df = pd.merge(df, sentiment_df, on='Date', how='inner')

    # Calculate daily stock returns
    merged_df['daily_return'] = merged_df['Close'].pct_change()

    # Drop rows with NaN values (which occur due to pct_change)
    merged_df.dropna(subset=['daily_return', 'sentiment'], inplace=True)

    # Plot sentiment, closing price, and daily return
    sns.lineplot(data=merged_df, x='Date', y='Close', ax=axes[i], label='Closing Price', color='blue')
    sns.lineplot(data=merged_df, x='Date', y='sentiment', ax=axes[i], label='Sentiment', color='green')
    # sns.lineplot(data=merged_df, x='Date', y='daily_return', ax=axes[i], label='Daily Return', color='orange')

    axes[i].set_title(f'{stock} - Sentiment, Daily Return, and Closing Price')
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Values')
    axes[i].legend()
    axes[i].grid(True)

# Remove the last empty subplot (6th subplot)
fig.delaxes(axes[-1])

# Adjust layout
plt.tight_layout()
plt.show()
```
