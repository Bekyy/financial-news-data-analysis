```python
import pandas as pd

# Initialize empty lists to store DataFrames
close_price_dfs = []
close_price_rtrn_dfs = []

# Iterate over each DataFrame and prepare the 'Close' prices and 'Weekly Return'
for stock_name, df in dataframes.items():
    # Ensure the 'Date' column is set as the index if not already
    if 'Date' in df.columns:
        df = df.set_index('Date')

    # Select the 'Close' column and rename it to the stock name
    close_price_df = df[['Close']].rename(columns={'Close': stock_name})
    # Select the 'Weekly Return' column and rename it to the stock name
    close_price_rtrn_df = df[['Weekly Return']].rename(columns={'Weekly Return': stock_name})

    # Append the DataFrame to the lists
    close_price_dfs.append(close_price_df)
    close_price_rtrn_dfs.append(close_price_rtrn_df)

# Concatenate all the DataFrames along the columns
combined_close_prices = pd.concat(close_price_dfs, axis=1)
combined_ret_prices = pd.concat(close_price_rtrn_dfs, axis=1)

# Reset the index to ensure 'Date' is a column again
combined_close_prices = combined_close_prices.reset_index()
combined_ret_prices = combined_ret_prices.reset_index()

# Convert 'Date' column to datetime if it's not already
combined_close_prices['Date'] = pd.to_datetime(combined_close_prices['Date'])
combined_ret_prices['Date'] = pd.to_datetime(combined_ret_prices['Date'])

# Filter the DataFrame for dates on or after '2014-01-01'
ech_close = combined_close_prices[combined_close_prices['Date'] >= '2014-01-01']
tech_rets = combined_close_prices[combined_ret_prices['Date'] >= '2014-01-01']

ech_close = ech_close.set_index('Date')
tech_rets = tech_rets.set_index('Date')
# Display the first few rows of the filtered DataFrame
#tech_rets.head()

# Display the first few rows of the combined DataFrames
print("Combined Closing Prices:")
print(ech_close.head())

print("\nCombined Weekly Returns:")
print(tech_rets.head())
```

```python
import cvxpy as cp

class PortfolioOptimizer:
    def __init__(self, dataframes):
        self.dataframes = dataframes

    def prepare_data(self, tickers, start_date, end_date):
        # Combine the 'Close' prices from all selected DataFrames
        data = pd.concat([self.dataframes[ticker][['Close']] for ticker in tickers], axis=1)
        data.columns = tickers

        # Filter data within the specified date range
        data = data[(data.index >= start_date) & (data.index <= end_date)]
        return data

    def calculate_portfolio_weights(self, tickers, start_date, end_date):
        data = self.prepare_data(tickers, start_date, end_date)

        # Calculate returns
        returns = data.pct_change().dropna()

        # Mean returns and covariance matrix
        mu = returns.mean().values
        cov = returns.cov().values

        # Define optimization variables
        n = len(mu)
        weights = cp.Variable(n)

        # Define the objective function: maximize return subject to volatility constraints
        portfolio_return = cp.sum(cp.multiply(mu, weights))
        portfolio_volatility = cp.quad_form(weights, cov)
        risk_free_rate = 0.01

        # Optimize for return with a volatility constraint
        objective = cp.Maximize(portfolio_return)
        constraints = [
            cp.sum(weights) == 1,        # Weights must sum to 1
            weights >= 0,                # No short selling
            portfolio_volatility <= 0.05 # Volatility constraint (adjust as needed)
        ]

        # Solve the problem
        prob = cp.Problem(objective, constraints)
        prob.solve()

        # Extract and return weights
        weights_opt = weights.value
        weights_dict = dict(zip(tickers, weights_opt))
        return weights_dict

    def calculate_portfolio_performance(self, tickers, start_date, end_date):
        data = self.prepare_data(tickers, start_date, end_date)

        # Calculate returns
        returns = data.pct_change().dropna()

        # Mean returns and covariance matrix
        mu = returns.mean().values
        cov = returns.cov().values

        # Define optimization variables
        n = len(mu)
        weights = cp.Variable(n)

        # Define the objective function: maximize return subject to volatility constraints
        portfolio_return = cp.sum(cp.multiply(mu, weights))
        portfolio_volatility = cp.quad_form(weights, cov)
        risk_free_rate = 0.01

        # Optimize for return with a volatility constraint
        objective = cp.Maximize(portfolio_return)
        constraints = [
            cp.sum(weights) == 1,        # Weights must sum to 1
            weights >= 0,                # No short selling
            portfolio_volatility <= 0.05 # Volatility constraint (adjust as needed)
        ]

        # Solve the problem
        prob = cp.Problem(objective, constraints)
        prob.solve()

        # Calculate portfolio performance metrics
        weights_opt = weights.value
        portfolio_return = np.dot(mu, weights_opt)
        portfolio_volatility = np.sqrt(np.dot(weights_opt.T, np.dot(cov, weights_opt)))
        sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_volatility

        return portfolio_return, portfolio_volatility, sharpe_ratio

# Example usage
dataframes = {
    'AAPL': weekly_data,
    'NVDA': weekly_NVDA_data,
    'META': weekly_META_data,
    'AMZN': weekly_AMZN_data,
    'GOOG': weekly_GOOG_data,
    'TSLA': weekly_TSLA_data,
    'MSFT': weekly_MSFT_data
}

optimizer = PortfolioOptimizer(dataframes)

# Define the tickers and date range
tickers = ['AAPL', 'NVDA', 'META','AMZN','GOOG','TSLA','MSFT'] # Adjust as needed
start_date = '2014-01-01'
end_date = '2023-01-01'

# Calculate portfolio weights and performance
weights = optimizer.calculate_portfolio_weights(tickers, start_date, end_date)
performance = optimizer.calculate_portfolio_performance(tickers, start_date, end_date)

# Print results
print("Optimal Weights:")
print(weights)

print("\nPortfolio Performance:")
print(f"Return: {performance[0]:.2f}")
print(f"Volatility: {performance[1]:.2f}")
print(f"Sharpe Ratio: {performance[2]:.2f}")
```
